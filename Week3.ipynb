```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Ques 1:

```python
x_values = [0,1.5,2.0,3.5,4.0,3.5]
y_values = [2.0,2.5,3.0,2.0,2.0,2.5]
sns.lineplot(x=x_values, y=y_values)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Zig-Zag Line Plot')
plt.legend()
plt.show()
```

## Ques# 2

```python
x_values1 = [1, 2, 3, 4, 5]
y_values1 = [1, 2, 1, 3, 2]
x_values2 = [3, 4, 5, 6, 7]
y_values2 = [3, 4, 3, 5, 4]
fig, ax = plt.subplots()
ax.plot(x_values1, y_values1, label='Line 1')
ax.plot(x_values2, y_values2, label='Line 2')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_title('Zig-Zag Lines Intersecting')
```

## Ques# 3

```python
x_values1 = [1, 2, 3, 4, 5]
y_values1 = [1, 3, 1, 3, 1]
x_values2 = [1, 2, 3, 4, 5]
y_values2 = [2, 4, 2, 4, 2]
fig, ax = plt.subplots()
ax.plot(x_values1, y_values1, marker='o', markersize=10, label='Line 1')
ax.plot(x_values2, y_values2, marker='o', markersize=10, label='Line 2')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_title('Zig-Zag Lines with marker')
ax.grid(True)
ax.legend()
plt.show()
```
